```python
def rolling_beta_regression(df, window=5, y_col='ret', x_col='MKT'):
    """
    高效的滚动回归计算beta系数
    假设数据已按时间排序，且已按id分组排序
    """
    # 确保按id分组
    grouped = df.groupby('id', group_keys=False)
    
    # 定义滚动回归函数
    def calculate_rolling_beta(group):
        # 确保按时间排序
        group = group.sort_values('time')
        
        # 提取y和x
        y = group[y_col].values
        x = group[x_col].values
        
        # 初始化结果数组
        n = len(y)
        betas = np.full(n, np.nan)
        
        # 滚动计算beta
        for i in range(window - 1, n):
            # 获取窗口数据
            y_window = y[i-window+1:i+1]
            x_window = x[i-window+1:i+1]
            
            # 计算beta: cov(x,y) / var(x)
            x_mean = np.mean(x_window)
            y_mean = np.mean(y_window)
            
            numerator = np.sum((x_window - x_mean) * (y_window - y_mean))
            denominator = np.sum((x_window - x_mean) ** 2)
            
            # 防止除零错误
            if denominator != 0:
                betas[i] = numerator / denominator
        
        # 返回结果
        return pd.DataFrame({
            'time': group['time'],
            'id': group['id'].iloc[0],
            'beta': betas
        })
    
    # 并行计算每个id的滚动beta
    results = []
    for _, group in grouped:
        results.append(calculate_rolling_beta(group))
    
    return pd.concat(results, ignore_index=True)

# 直接使用
betas_df = rolling_beta_regression(df2, window=5, y_col='ret', x_col='MKT')

# 显示结果
print(betas_df.tail(20))
```

```python
def rolling_multi_factor_betas(df, window=5, y_col='ret'):
    """
    计算所有因子的滚动beta
    """
    # factors = ['MKT', 'SMB', 'HML', 'RMW', 'CMA']
    factors = ["MKT"]
    results = []
    
    for factor in factors:
        factor_betas = rolling_beta_regression(df, window=window, x_col=factor)
        factor_betas.rename(columns={'beta': f'beta_{factor}'}, inplace=True)
        
        if results:
            # 合并到已有结果
            results[-1] = pd.merge(results[-1], factor_betas, on=['time', 'id'])
        else:
            results.append(factor_betas)
    
    return results[0] if results else pd.DataFrame()

df_capm["ret"] = df_capm["ret"]
df_capm["MKT"] = df_capm["MKT"]
# 计算所有因子beta
all_betas = rolling_multi_factor_betas(df_capm, window=window)
print(all_betas.head(10))
```

```python
def rolling_multi_residuals(df, window=5, y_col='ret', x_cols=['MKT', "SMB", "HML"]):
    """
    多元滚动回归计算残差
    使用矩阵运算提高效率
    """
    results = []
    
    for id_name, group in df.groupby('id'):
        # 确保按时间排序
        group = group.sort_values('time')
        
        # 提取数据
        dates = group['time'].values
        y = group[y_col].values
        X = group[x_cols].values
        n = len(y)
        
        # 添加常数项
        X_with_const = np.column_stack([np.ones(n), X])
        
        # 初始化结果数组
        residuals = np.full(n, np.nan)
        betas_matrix = np.full((n, len(x_cols)), np.nan)
        
        # 滚动窗口计算
        for i in range(window - 1, n):
            start, end = i - window + 1, i + 1
            
            # 窗口数据
            X_window = X_with_const[start:end]
            y_window = y[start:end]
            
            try:
                # 最小二乘解: beta = (X'X)^(-1)X'y
                XtX = X_window.T @ X_window
                Xty = X_window.T @ y_window
                
                # 检查矩阵是否可逆
                if np.linalg.matrix_rank(XtX) == XtX.shape[0]:
                    beta = np.linalg.inv(XtX) @ Xty
                    
                    # 保存系数（不包括常数项）
                    betas_matrix[i, :] = beta[1:]
                    
                    # 计算当前期的残差
                    y_pred = X_with_const[i] @ beta
                    residuals[i] = y[i] - y_pred
                    
            except np.linalg.LinAlgError:
                # 矩阵不可逆，跳过
                continue
        
        # 创建结果DataFrame
        df_result = pd.DataFrame({
            'time': dates,
            'id': id_name,
            'residual': residuals
        })
        
        # 添加每个因子的beta
        for j, col in enumerate(x_cols):
            df_result[f'beta_{col}'] = betas_matrix[:, j]
        
        results.append(df_result)
    
    return pd.concat(results, ignore_index=True)

# 使用多元回归
multi_residuals = rolling_multi_residuals(
    df_ff3, 
    window=window, 
    x_cols=['MKT', 'SMB', 'HML']
)
print(multi_residuals.head(10))
```

```python
# 如果你的数据格式是"2013-01"这样的字符串
def calculate_survival_time_string(df, time_col='time', id_col='id'):
    """
    适用于时间格式为"YYYY-MM"字符串的情况
    """
    # 确保按时间排序
    df = df.copy().sort_values([id_col, time_col]).reset_index(drop=True)
    
    # 将时间字符串转换为年月整数便于计算
    df['year_month'] = df[time_col].astype(str).apply(
        lambda x: int(x[:4]) * 12 + int(x[5:7]) - 1  # 转换为月份总数
    )
    
    # 计算每个股票的首次出现月份
    first_month = df.groupby(id_col)['year_month'].min().reset_index()
    first_month.columns = [id_col, 'first_month']
    
    # 合并
    df_with_first = pd.merge(df, first_month, on=id_col)
    
    # 计算存续月份数
    df_with_first['survival_months'] = df_with_first['year_month'] - df_with_first['first_month']
    
    # 返回结果（删除辅助列）
    return df_with_first[[time_col, id_col, 'survival_months']]

# df_age_month = calculate_survival_time_string(ret3)
```

```python
df_ta_copy = df_ta.copy()

def fill_quarter_to_month(df):
    """
    将季度数据填充到月度，确保所有月份都有数据
    """
    df = df.copy()
    
    # 转换时间格式
    df['time'] = df['time'].dt.to_timestamp()
    
    results = []
    
    # 对每个股票单独处理
    for stkcd, group in df.groupby('Stkcd'):
        # 按时间排序
        group = group.sort_values('time')
        
        # 生成完整的月度时间范围
        min_date = group['time'].min()
        max_date = group['time'].max()
        
        # 创建完整的月度日期序列
        all_months = pd.date_range(
            start=min_date,
            end=max_date,
            freq='MS'  # 月初
        )
        
        # 创建包含所有月份的DataFrame
        full_df = pd.DataFrame({
            'Stkcd': stkcd,
            'time': all_months
        })
        
        # 合并原始数据
        merged = pd.merge(
            full_df,
            group,
            on=['Stkcd', 'time'],
            how='left'
        )
        
        # 向前填充所有数值列
        numeric_cols = merged.select_dtypes(include=[np.number]).columns
        merged[numeric_cols] = merged[numeric_cols].ffill()
        
        results.append(merged)
    
    # 合并所有股票
    result_df = pd.concat(results, ignore_index=True)
    
    # 按股票和时间排序
    result_df = result_df.sort_values(['Stkcd', 'time']).reset_index(drop=True)
    
    return result_df

# 使用推荐方案
df_ta_copy = fill_quarter_to_month(df_ta_copy)

# 计算 total_accrual
df_ta_copy["total_accruals"] = df_ta_copy["current_asset_chg"] - df_ta_copy["cash_chg"] - df_ta_copy["current_liability_chg"] - df_ta_copy["depreciation_TTM"] - df_ta_copy["short_term_debt_chg"] - df_ta_copy["depreciation_TTM"]


df_total_accruals = df_ta_copy[["Stkcd", "time", "total_accruals"]]
df_total_accruals["time"] = df_total_accruals["time"].dt.to_period("M")

df_ta_copy.head()
```
