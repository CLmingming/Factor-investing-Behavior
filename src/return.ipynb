```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
root_path = Path('../')
monthly_return_path = root_path / 'data' / 'monthly_returns.pkl'
monthly_return = pd.read_pickle(monthly_return_path).reset_index()
monthly_return = monthly_return.melt(
    id_vars=["date"],
    var_name="asset",
    value_name="return"
)
monthly_return['date'] = pd.to_datetime(monthly_return['date'])
monthly_return = monthly_return[(monthly_return['date'] >= '2014-01-01') & (monthly_return['date'] <= '2024-12-31')]
monthly_return
```
